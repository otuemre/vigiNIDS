# Modelling

## Get the Data Ready

```python
# Imports
import pandas as pd

train_df = pd.read_csv("../datasets/UNSW_NB15_preprocessed_training-set.csv")
test_df = pd.read_csv("../datasets/UNSW_NB15_preprocessed_testing-set.csv")
```

```python
train_df.head()
```

```python
train_df.columns, len(train_df.columns)
```

```python
test_df.head()
```

```python
test_df.columns, len(test_df.columns)
```

```python
# Create X and y and drop id
X_train = train_df.drop(columns=["id", "label", "attack_cat"])
y_train = train_df["label"]

X_test = test_df.drop(columns=["id", "label", "attack_cat"])
y_test = test_df["label"]
```

**Note**: During my experiments, I have reliazed the `attack_cat` is highly relevant for `label` which makes the model a perfect model with %100. It shouldn't be the case as this is a real-world dataset. Therefore, I have decided to drop `attack_cat` for a better model. My aim to have %90-%95 accuracy!

```python
X_train.columns
```

## XGBoost

```python
y_train
```

```python
from xgboost import XGBClassifier
import xgboost as xgb

xgb_model = XGBClassifier(tree_method="hist", device="cuda", verbosity=2)
xgb_model.fit(X_train, y_train)
```

```python
xgb_model.score(X_test, y_test)
```

```python
import matplotlib.pyplot as plt

# Plot Feature Importance
xgb.plot_importance(xgb_model, max_num_features=10)

# Save the image
plt.savefig("../images/xgboost_base_feature_importance.png")

# Show the plot
plt.show()
```

```python
# Import our custom function from ./src/utils/print_unique_values.py
import sys
import os
from pathlib import Path

# Get the current notebook's directory
notebook_dir = Path.cwd()

# Add the src directory to the sys.path
sys.path.append(str(notebook_dir.parent))  # Go up one level from notebooks

# Import the function
from utils.evaluate_classification import evaluate_classification
from utils.plot_auc_roc import plot_auc_roc
from utils.plot_confusion_matrix import plot_confusion_matrix
```

```python
# Predict labels and probabilities
y_pred = xgb_model.predict(X_test)
y_proba = xgb_model.predict_proba(X_test)[:, 1]
```

```python
# Get evaluation metrics
xgb_metrics = evaluate_classification(y_test, y_pred, "XGBoost")
print("Evaluation Metrics:")
xgb_metrics
```

```python
# Plot the confusion matrix
plot_confusion_matrix(y_test, y_pred, class_names=["Normal", "Attack"])
```

**Insigth**: It seems that the model is confused when the truth label is `normal`. It is probably the imbalance of train_df. As we see on EDA section, there were a lot more attack than normal. However, it might be beneficial as this system requires high attention. Instead of labelling an `attack` as `normal`, labelling `normal` as `attack` is better. Even though `recall` is more important for these type of systems, I will try to improve `precision` as well.

```python
# Plot AUC-ROC
xgb_auc = plot_auc_roc(y_test, y_proba)
print(f"AUC: {xgb_auc:.4f}")
```

## TabNet

```python
# Convert boolean to int
train_bool_columns = X_train.select_dtypes(include=["bool"]).columns
test_bool_columns = X_test.select_dtypes(include=["bool"]).columns

X_train[train_bool_columns] = X_train[train_bool_columns].astype(int)
X_test[test_bool_columns] = X_test[test_bool_columns].astype(int)
```

```python
X_train.select_dtypes(include=["bool"]).columns
```

```python
X_test.select_dtypes(include=["bool"]).columns
```

```python
from pytorch_tabnet.tab_model import TabNetClassifier
import numpy as np

# Convert data to tensors
X_train_np = X_train.values.astype("float32")  # Ensure float32 dtype
y_train_np = y_train.values.astype("int64")  # Ensure int64 dtype

X_test_np = X_test.values.astype("float32")
y_test_np = y_test.values.astype("int64")

tabnet_model = TabNetClassifier(device_name="cuda")
tabnet_model.fit(X_train_np, y_train_np, eval_set=[(X_test_np, y_test_np)])
```

```python
y_pred = tabnet_model.predict(X_test_np)
y_proba = tabnet_model.predict_proba(X_test_np)[:, 1]
```

```python
# Get evaluation metrics
tabnet_metrics = evaluate_classification(y_test_np, y_pred, "TabNet")
print("Evaluation Metrics:")
tabnet_metrics
```

```python
# Plot the confusion matrix
plot_confusion_matrix(y_test_np, y_pred, class_names=["Normal", "Attack"])
```

```python
# Plot AUC-ROC
tabnet_auc = plot_auc_roc(y_test_np, y_proba)
print(f"AUC: {tabnet_auc:.4f}")
```

**Insight**: The `TabNet` is better in terms of accuracy. However, the recall and AUC lowered a bit.

**Results DataFrame**

```python
import pandas as pd

# Add AUC Scores
xgb_metrics["AUC Score"] = xgb_auc
lgbm_metrics["AUC Score"] = lgbm_auc
catboost_metrics["AUC Score"] = catboost_auc
tabnet_metrics["AUC Score"] = tabnet_auc

model_res = [
    xgb_metrics,
    lgbm_metrics,
    catboost_metrics,
    tabnet_metrics
]

# Convert results into a DataFrame
results_df = pd.DataFrame(model_res)

# Display the results
results_df
```

**Insight**: The evaluation results indicate that all models performed well on the classification task, with high accuracy and AUC scores. LightGBM and CatBoost achieved the highest accuracy at **87.71%**, while XGBoost and TabNet followed closely. In terms of **Recall**, LightGBM and CatBoost demonstrated the best performance, capturing **98.53%** of the positive cases. 

However, **Precision** shows a slight variation, with XGBoost and TabNet maintaining slightly better balance between false positives and true positives. The **AUC scores** further validate the strong predictive power of all models, with LightGBM and CatBoost achieving the highest at **0.9862**.

Overall, LightGBM and CatBoost stand out as the top-performing models, making them ideal candidates for further tuning and optimization.
