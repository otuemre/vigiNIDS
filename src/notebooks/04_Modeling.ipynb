# Modelling

## Get the Data Ready

```python
# Imports
import pandas as pd

train_df = pd.read_csv("../datasets/UNSW_NB15_preprocessed_training-set.csv")
test_df = pd.read_csv("../datasets/UNSW_NB15_preprocessed_testing-set.csv")
```

```python
test_df.head()
```

```python
test_df.columns, len(test_df.columns)
```

```python
# Create X and y and drop id
X_train = train_df.drop(columns=["id", "label", "attack_cat"])
y_train = train_df["label"]

X_test = test_df.drop(columns=["id", "label", "attack_cat"])
y_test = test_df["label"]
```

**Note**: During my experiments, I have reliazed the `attack_cat` is highly relevant for `label` which makes the model a perfect model with %100. It shouldn't be the case as this is a real-world dataset. Therefore, I have decided to drop `attack_cat` for a better model. My aim to have %90-%95 accuracy!

```python
X_train.columns
```

## XGBoost

```python
y_train
```

```python
from xgboost import XGBClassifier
import xgboost as xgb

xgb_model = XGBClassifier(tree_method="hist", device="cuda", verbosity=2)
xgb_model.fit(X_train, y_train)
```

```python
xgb_model.score(X_test, y_test)
```

```python
import matplotlib.pyplot as plt

# Plot Feature Importance
xgb.plot_importance(xgb_model, max_num_features=10)

# Save the image
plt.savefig("../images/xgboost_base_feature_importance.png")

# Show the plot
plt.show()
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(xgb_model, X_train, y_train, cv=5, scoring='accuracy')
print(f"Cross-validation scores: {scores}")
print(f"Mean CV Accuracy: {scores.mean()}")
```
