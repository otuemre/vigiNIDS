# Modelling

## Get the Data Ready

```python
# Imports
import pandas as pd

train_df = pd.read_csv("../datasets/UNSW_NB15_preprocessed_training-set.csv")
test_df = pd.read_csv("../datasets/UNSW_NB15_preprocessed_testing-set.csv")
```

```python
train_df.head()
```

```python
train_df.columns, len(train_df.columns)
```

```python
test_df.head()
```

```python
test_df.columns, len(test_df.columns)
```

```python
# Create X and y and drop id
X_train = train_df.drop(columns=["id", "label", "attack_cat"])
y_train = train_df["label"]

X_test = test_df.drop(columns=["id", "label", "attack_cat"])
y_test = test_df["label"]
```

**Note**: During my experiments, I have reliazed the `attack_cat` is highly relevant for `label` which makes the model a perfect model with %100. It shouldn't be the case as this is a real-world dataset. Therefore, I have decided to drop `attack_cat` for a better model. My aim to have %90-%95 accuracy!

```python
X_train.columns
```

## XGBoost

```python
y_train
```

```python
from xgboost import XGBClassifier
import xgboost as xgb

xgb_model = XGBClassifier(tree_method="hist", device="cuda", verbosity=2)
xgb_model.fit(X_train, y_train)
```

```python
xgb_model.score(X_test, y_test)
```

```python
import matplotlib.pyplot as plt

# Plot Feature Importance
xgb.plot_importance(xgb_model, max_num_features=10)

# Save the image
plt.savefig("../images/xgboost_base_feature_importance.png")

# Show the plot
plt.show()
```

```python
# Import our custom function from ./src/utils/print_unique_values.py
import sys
import os
from pathlib import Path

# Get the current notebook's directory
notebook_dir = Path.cwd()

# Add the src directory to the sys.path
sys.path.append(str(notebook_dir.parent))  # Go up one level from notebooks

# Import the function
from utils.evaluate_classification import evaluate_classification
from utils.plot_auc_roc import plot_auc_roc
from utils.plot_confusion_matrix import plot_confusion_matrix
```

```python
# Predict labels and probabilities
y_pred = xgb_model.predict(X_test)
y_proba = xgb_model.predict_proba(X_test)[:, 1]
```

```python
# Get evaluation metrics
metrics = evaluate_classification(y_test, y_pred)
print("Evaluation Metrics:")
metrics
```

```python
# Plot the confusion matrix
plot_confusion_matrix(y_test, y_pred, class_names=["Normal", "Attack"])
```

**Insigth**: It seems that the model is confused when the truth label is `normal`. It is probably the imbalance of train_df. As we see on EDA section, there were a lot more attack than normal. However, it might be beneficial as this system requires high attention. Instead of labelling an `attack` as `normal`, labelling `normal` as `attack` is better. Even though `recall` is more important for these type of systems, I will try to improve `precision` as well.

```python
# Plot AUC-ROC
auc = plot_auc_roc(y_test, y_proba)
print(f"AUC: {auc:.4f}")
```
